# Checking for Stationarity 

```python
from statsmodels.tsa.stattools import adfuller
```

```python

def adf_test(dataset):
  dftest = adfuller(dataset, autolag = "AIC")
  print("1. ADF : ",dftest[0])
  print("2. P-Value : ", dftest[1])
  print("3. Num Of Lags : ", dftest[2])
  print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
  print("5. Critical Values :")
  for key, val in dftest[4].items():
      print("\t",key, ": ", val)
```

```python
adf_test(df["buying_rate"])
```

f p< 0.05 ; Data is stationary
if p>0.05; Data is not stationary

```python
n = int(len(df) * 0.8)
train = df[:n]
test=df[n:] 
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(train["buying_rate"],order=(5,1,1))
model=model.fit(disp=0)
model.summary()
```

```python
fc, se, conf = model.forecast(len(test))
```

```python
fc = pd.Series(fc, index=test['buying_rate'].index) 

lower = pd.Series(conf[:,0], index=test['buying_rate'].index) 
upper = pd.Series(conf[:,1], index=test['buying_rate'].index) 
```

```python
plt.figure(figsize=(16,8))
plt.plot(test['buying_rate'], label="actual", color="blue")
plt.plot(fc, label = "forecast", color="green")
#plt.plot(df['buying_rate'], label = "a", color="red")
plt.fill_between(lower.index, lower, upper, color="k", alpha=0.1)
plt.title("Forecast vs Actual")
plt.legend(loc="upper left")
```
